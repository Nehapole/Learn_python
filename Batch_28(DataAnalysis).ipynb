```python
import matplotlib.pyplot as plt 
import numpy as np

ypoint= np.array([3,8,1,12,8])
plt.plot(ypoint,'*:r',ms=20)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

ypoint= np.array([3,8,1,12,8])
plt.plot(ypoint,'*:r',ms=20,mfc='c')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

ypoint= np.array([3,8,1,12,8])
plt.plot(ypoint,'*:r',ms=20,mfc='#4CAF50')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

ypoint= np.array([3,8,1,12,8])
plt.plot(ypoint,linestyle='dotted',ms=20,mfc='c')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data= sns.load_dataset("planets")
sns.lineplot(x="number", y="year", data=data)
sns.set_style("dark")

plt.xlim()

plt.title("Number of planets in the given year")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data= sns.load_dataset("planets")
sns.lineplot(x="number", y="year", data=data)
plt.figure(figsize=(4,6))
sns.despine(left=True)



plt.title("Number of planets in the given year")
plt.show()
```
